```python
DATA_ROOT = '../data/ms_coco2017'
TRAIN_IMG = DATA_ROOT + '/train2017'
VALID_IMG = DATA_ROOT + '/val2017'
TRAIN_LBL = DATA_ROOT + '/annotations/panoptic_train2017.json'
VALID_LBL = DATA_ROOT + '/annotations/panoptic_val2017.json'
```

### Label as a json file

using `json.load(filename)` to load label file. 

- info
- licenses
- images : list of dictionary
    - license : Int
    - file_name : String, like '000000397133.jpg'
    - coco_url  : String, like 'http://images.cocodataset.org/val2017/000000397133.jpg'
    - height    : Int
    - width     : Int
    - data_captured : String, like '2013-11-14 17:02:52'
    - flickr_url : String, like 'http://farm7.staticflickr.com/6116/6255196340_da26cf2c9e_z.jpg'
    - id : Int, like 397133
- annotations : list of dictionary
    - segments_info : list of dictionary
        - id : Int, like 3226956
        - category_id : Int, like 1
        - iscrowd : Int, like 0
        - bbox : list, like [413, 158, 53, 138] --> xmin, ymin, width , height
        - area : Int, like 2840
    - file_name : String, like '000000000139.png'
    - image_id : Int, like 139
- categories : list of dictonary
    - supercategory : String, like person
    - isthing : Int, like 1
    - id : Int, like 1
    - name : String, like person
